```python
#Using histogram to represent relationship between all features
df.hist(figsize=(100,80))
plt.show()
```

```python
#countries affected by terrorist attacks
df['country'].value_counts()
plt.figure(figsize=(20,10))
sns.barplot(x=df['country_txt'].value_counts()[:20].index,y=df['country_txt'].value_counts()[:20].values)
plt.title('TOP COUNTRIES THAT GOT ATTACKED')
plt.xticks(rotation=90)
plt.xlabel('COUNTRIES')
plt.ylabel('NUMBER OF ATTACKS')
plt.show()
```

```python
# Regions affected by terrorist attacks
df['region_txt'].value_counts()
plt.figure(figsize=(20,10))
sns.countplot(x='region_txt',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
df['city'].value_counts()
```

```python
# Number of attacks from 1970 to 2017
plt.subplots(figsize=(20,10))
sns.countplot(x='iyear',data=df,palette='tab10')
plt.xticks(rotation=90)
plt.title('TERROR ATTACKS FROM 1970 TO 2017')
plt.show()
```

```python
#Type of attack used by terrorists
plt.figure(figsize=(10,10))
sns.countplot(x=df['attacktype1_txt'],order=df['attacktype1_txt'].value_counts().index,palette='rainbow')
plt.xticks(rotation=90)
plt.xlabel('ATTACK TYPE')
plt.ylabel('NUMBER OF ATTACKS')
plt.show()
```

```python
#Attacks target
plt.figure(figsize=(20,10))
sns.countplot(x=df['targtype1_txt'],order=df['targtype1_txt'].value_counts().index,palette='rainbow')
plt.xticks(rotation=90)
plt.xlabel('TARGETS')
plt.ylabel('NUMBER OF TARGETS')
plt.show()
```

```python
#Name of group attacks terrorist
plt.figure(figsize=(15,10))
sns.barplot(x=df['gname'].value_counts()[:20].index,y=df['gname'].value_counts()[:20].values,palette='rainbow')
plt.xticks(rotation=90)
plt.title('TOTAL GROUP ATTACKS')
plt.xlabel('GROUP NAME')
plt.ylabel('TOTAL NUMBER OF ATTACKS')
plt.show()
```

```python
#cities affected by terrorist
df['city'].value_counts()
plt.figure(figsize=(25,10))
sns.barplot(x=df['city'].value_counts()[:20].index,y=df['city'].value_counts()[:20].values)
plt.xticks(rotation=90)
plt.title('TOP CITIES GOT ATTACKED')
plt.xlabel('CITIES')
plt.ylabel('NUMBER OF ATTACKS')
plt.show()
```

```python
#weapons used for supporting the attack type visual
df['weaptype1_txt'].value_counts()
plt.figure(figsize=(15,10))
sns.barplot(x=df['weaptype1_txt'].value_counts()[:20].index,y=df['weaptype1_txt'].value_counts()[:20].values)
plt.xticks(rotation=90)
plt.title('WEAPONS USAGE')
plt.xlabel('WEAPONS USED')
plt.ylabel('NUMBER OF TIMES')
plt.show()
```
