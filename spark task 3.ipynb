```python
#Data visualization
#plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=data1)
plt.title("category vs sub-category")
plt.xlabel("sub-category")
plt.ylabel("category")
plt.xticks(rotation=45)
plt.show()
```

```python
data1.hist(bins=50,figsize=(20,15))
plt.show();
```

```python
#count total repeatable states
data1['State'].value_counts()
```

```python
plt.figure(figsize=(15,15))
sns.countplot(x=data1['State'])
plt.title("STATES")
plt.xticks(rotation=90)
plt.show()
```

```python
#profit
fig,axes= plt.subplots(1,2,figsize=(14,5))
fig.suptitle("total profit vs sales")
sns.barplot(data=data1.groupby('Sub-Category')['Sales','Profit'].agg(sum),x='Sales',y='Profit',ax=axes[1])
data1.groupby('Sub-Category')['Sales','Profit'].agg(sum).plot(kind='bar',ax=axes[0])
plt.xticks(rotation=90)
plt.show()
```

```python
fig,axes= plt.subplots(1,2,figsize=(14,5))
fig.suptitle("total sales vs Quantity")
sns.barplot(data=data1.groupby('Sub-Category')['Sales','Quantity'].agg(sum),x='Sales',y='Quantity',ax=axes[1])
data1.groupby('Sub-Category')['Sales','Quantity'].agg(sum).plot(kind='bar',ax=axes[0])
plt.xticks(rotation=90)
plt.show()
```

```python
fig,axes= plt.subplots(1,2,figsize=(14,5))
data1.groupby('Sub-Category')['Profit','Quantity'].agg(sum).plot(kind='bar',ax=axes[0]).set_title("Quantity & profit relation based on sub category")
data1.groupby('Sub-Category')['Profit','Discount'].agg(sum).plot(kind='bar',ax=axes[1]).set_title("Discount & profit relation based on sub category")
plt.xticks(rotation=90)
plt.show()
```

```python
fig,axes= plt.subplots(2,2,figsize=(14,5))
fig.suptitle("Distribution plots",fontsize=15)
sns.distplot(data1['Sales'],ax=axes[0,0])
sns.distplot(data1['Profit'],ax=axes[0,1])
sns.distplot(data1['Discount'],ax=axes[1,0])
sns.distplot(data1['Quantity'],ax=axes[1,1])
plt.show()
```

```python
fig,ax=plt.subplots(1,1,figsize=(14,5))
sns.countplot(data1['Quantity'],hue=data1['Region'])
plt.show()
```

```python
fig,axes= plt.subplots(2,2,figsize=(14,5))
fig.suptitle("Sales with different shipping modes and segments",fontsize=16)
sns.barplot(data1['Ship Mode'],data1['Sales'],ax=axes[0,0])
sns.lineplot(data1['Ship Mode'],data1['Sales'],ax=axes[0,1])
sns.barplot(data1['Segment'],data1['Sales'],ax=axes[1,0])
sns.lineplot(data1['Segment'],data1['Sales'],ax=axes[1,1])
plt.show()
```
